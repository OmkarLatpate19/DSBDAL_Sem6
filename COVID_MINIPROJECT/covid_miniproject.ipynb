# Data Preprocessing

```python
df["First Dose Administered"].fillna(value = df["First Dose Administered"].mean(axis = 0) , inplace=True)
```

```python
df["Total Individuals Registered"].fillna(value = df["Total Individuals Registered"].mean(axis = 0) , inplace=True)
```

```python
df["Total Sessions Conducted"].fillna(value = df["Total Sessions Conducted"].mean(axis = 0) , inplace=True)
```

```python
df["Second Dose Administered"].fillna(value = df["Second Dose Administered"].mean(axis = 0) , inplace=True)
```

```python
df["Male(Individuals Vaccinated)"].fillna(value = df["Male(Individuals Vaccinated)"].mean(axis = 0) , inplace=True)
```

```python
df["Female(Individuals Vaccinated)"].fillna(value = df["Female(Individuals Vaccinated)"].mean(axis = 0) , inplace=True)
```

```python
df["Transgender(Individuals Vaccinated)"].fillna(value = df["Transgender(Individuals Vaccinated)"].mean(axis = 0) , inplace=True)
```

```python
df["Total Covaxin Administered"].fillna(value = df["Total Covaxin Administered"].mean(axis = 0) , inplace=True)
```

```python
df["Total CoviShield Administered"].fillna(value = df["Total CoviShield Administered"].mean(axis = 0) , inplace=True)
```

```python
df["Total Individuals Vaccinated"].fillna(value = df["Total Individuals Vaccinated"].mean(axis = 0) , inplace=True)
```

```python
df.isnull().sum()
```

# Number of persons state wise vaccinated for first dose in India

```python
first_dose = df.groupby('State')[['First Dose Administered']].sum()
first_dose
```

# Number of persons state wise vaccinated for second dose in India

```python
first_dose = df.groupby('State')[['Second Dose Administered']].sum()
first_dose
```
